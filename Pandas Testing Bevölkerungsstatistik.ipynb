```python
import pandas as pd
dataset = pd.read_csv("https://data.stadt-zuerich.ch/dataset/a9febb98-23dd-4438-95c6-9cad417d0291/resource/a90b0184-c809-48b4-87ca-91e5811da885/download/bev325od3250.csv")
```

```python
print(dataset.head())
```

```python
# fill_value to replace missing values (NaN)
pivdata = dataset.pivot_table(values="AnzBestWir", index="StichtagDatJahr", columns="StichtagDatMM", aggfunc=sum, fill_value="-")
print(pivdata)
```

```python
# pivot tables are indexed sorted tables and can be sliced with loc
pivdata.loc["2000":"2009", "3":"6"]
```

```python
max_year = dataset["StichtagDatJahr"].max()
print(max_year)
```

```python
data_year = dataset[dataset["StichtagDatJahr"]==max_year]
max_mon = data_year["StichtagDatMM"].max()
print(max_mon)
```

```python
data_mon = data_year[data_year["StichtagDatMM"]==max_mon]
print(data_mon["AnzBestWir"].sum())
```

```python
# margin = show totals
data_mon.pivot_table(values="AnzBestWir", index="HerkunftLang", columns="SexLang", aggfunc=sum, margins=True)
```

```python
data_mon.pivot_table(values="AnzBestWir", index=["KreisLang", "QuarLang"], columns=["HerkunftLang", "SexLang"], aggfunc=sum, margins=True)
```

```python
data_mon.pivot_table(values="AnzBestWir", index="AlterV20Kurz", columns="SexLang", aggfunc=sum, margins=True)
```

```python
# Use index to select easier with loc
newdata = dataset.reset_index()
newdata = dataset.set_index(["StichtagDatJahr", "StichtagDatMM"]).sort_index()
# this results in an hierarchical index
data_june_2003 = newdata.loc[2003, 6]
#print(data_june_2003.head())
# slicing is possible too by limiting from the first tupel to the last one
# and separated with a comma, columns can be selected (and sliced) too
data_10_years = newdata.loc[(1998, 1):(2018, 12), ["SexLang", "AlterV20Kurz", "HerkunftLang", "QuarLang", "AnzBestWir"]]
print(data_10_years)
```
